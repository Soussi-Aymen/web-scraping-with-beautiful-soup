```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
product = input("What product do  you want to search for ? ") 
```

```python
product = '+'.join(product.split())
```

```python
print(product)
```

```python
url = f"https://www.mytek.tn/catalogsearch/result/index/?cat=&product_list_dir=asc&product_list_limit=all&product_list_order=price&q={product}"
```

```python
page = requests.get(url).text
```

```python
doc = BeautifulSoup(page,"html.parser")
```

```python
print(doc)
```

```python
div = doc.find( "div", {"id":"amasty-shopby-product-list"})
```

```python
articles = div.find_all("tr",{"class":"item product product-item product-item-info"})
```

```python
prices=[]
disponibilities=[]
titles=[]
descriptions=[]
links=[]
```

```python
for article in articles:
    title = article.find("a",{"class":"product-item-link"}).text
    titles.append(title)
    link = article.find("a",{"class":"product-item-link"})["href"]
    links.append(link)
    description = article.find("div",{"class":"product description"}).text.replace(" En savoir plus","")
    descriptions.append(description)
    disponibility = article.find("div",{"itemprop":"availability" }).span.text
    disponibilities.append(disponibility)
    price = int(''.join(article.find("span",{"class":"price"}).text.replace(",","").replace("TND","").split()))/1000
    prices.append(price)
```

```python
titles
```

```python
df = pd.DataFrame({"Title":titles,"Disponibility":disponibilities,"Descritpion":descriptions,"Price":prices,"Link":links})
```
