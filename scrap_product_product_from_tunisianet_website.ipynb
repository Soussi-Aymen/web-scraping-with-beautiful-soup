```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
product = input("What product do  you want to search for ? ") 
```

```python
url = f"https://www.tunisianet.com.tn/recherche?controller=search&orderby=price&orderway=asc&s={product}&submit_search="
```

```python
page = requests.get(url).text
```

```python
doc = BeautifulSoup(page,"html.parser")
```

```python
print(doc)
```

```python
last_page = int(doc.find(class_="page-list clearfix").find_all("a")[-2].text)
```

```python
print(last_page)
```

```python
prices=[]
references=[]
titles=[]
descriptions=[]
links=[]
```

```python
for page in range(1,last_page+1):
    url = f"https://www.tunisianet.com.tn/recherche?controller=search&s={product}&submit_search=&page={page}&order=product.price.asc"

    
    page = requests.get(url).text
    doc = BeautifulSoup(page,"html.parser")
    div = doc.find( "div", {"class":"products product-thumbs row wb-product-list"})
    articles = div.find_all(class_="item-product col-xs-12")
    
    for article in articles:
        price = int(''.join(article.find(class_="product-price-and-shipping").span.string.replace(",","").replace(" DT","").split()))/1000
        prices.append(price)
        reference = article.find(class_="product-reference").text.replace("[","").replace("]","")
        references.append(reference)
        title = article.find(class_="h3 product-title").a.text
        titles.append(title)
        description = article.find(class_="listds").p.text
        if(description==''):
            description= " ".join(article.find( "div", {"class":"listds"}).a.text.replace("<span>","").replace("</span>","").split())
        descriptions.append(description)
        link = article.find(class_="h3 product-title").a["href"]
        links.append(link)
```

```python
df = pd.DataFrame({"Reference":references,"Title":titles,"Descritpion":descriptions,"Price":prices,"Link":links})
```
